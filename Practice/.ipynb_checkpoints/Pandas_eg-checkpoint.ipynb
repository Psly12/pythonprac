```python
df2=pandas.read_json('supermarkets.json')
df2.set_index('ID')
```

```python
from geopy.geocoders import Nominatim
nom=Nominatim(scheme='http')
```

```python
n=nom.geocode(list(df2.loc[df2.index[0],['Address','City','State','Country']]))
n.latitude
```

```python
df2["Address Str"]=df2['Address']+','+df2['City']+','+df2['State']+','+df2['Country']
df2
```

```python
df2['Coordinates']=df2['Address Str'].apply(nom.geocode)
```

```python
df2
```

```python
df2['Latitude'] = df2['Coordinates'].apply(lambda x: x.latitude if x!=None else None)
df2['Longitude'] = df2['Coordinates'].apply(lambda x: x.longitude if x!=None else None)
df2
```
